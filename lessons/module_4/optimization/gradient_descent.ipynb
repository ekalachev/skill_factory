# Градиентный спуск

$$x^{(n+1)} = x^{(n)} - \gamma \nabla f(x^{(n)})$$

- Здесь x(n) — точка проверки на n-ном шаге;
- $\nabla f(X^{(n)})$ — значение градиента в этой точке;
- $\gamma$ — темп обучения (learning rate).

```python
def f(x, y):
    return (1 - x) ** 2 + 100 * (y - x ** 2) ** 2


def grad(x, y):
    dx = 2 * x - 2 + 100 * (-4 * y * x + 4 * x ** 3)
    dy = 100 * (2 * y - 2 * x ** 2)
    return (dx, dy)


def dist(x1, x2):
    return (x1[0] - x2[0]) ** 2 + (x1[1] - x2[1]) ** 2
```

```python
x0 = (0, 2)
gamma = 1e-3
x_cur = x0

vals = []
coords = []
i = 0
while True:
    x_new = (x_cur[0] - gamma * grad(*x_cur)[0],
             x_cur[1] - gamma * grad(*x_cur)[1])
    if dist(x_new, x_cur) < 1e-8:
        break
    x_cur = x_new
    vals.append(f(*x_cur))
    coords.append(x_cur)
    i += 1
    print(f"iter={i}; x=({x_cur[0]:.2f}, {x_cur[1]:.2f});"
          f" f(x)={f(*x_cur):.2f}; grad f(x)=({grad(*x_cur)[0]:.2f}, {grad(*x_cur)[1]:.2f})")
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot(vals)
```

```python
n = 100  # number of discretization points along the x-axis
m = 100  # number of discretization points along the x-axis
a = -0.5
b = 2.  # extreme points in the x-axis
c = -1.5
d = 4.  # extreme points in the y-axis
X, Y = np.meshgrid(np.linspace(a, b, n), np.linspace(c, d, m))

Z = f(X, Y)

plt.xlim(a, b)
plt.ylim(c, d)

plt.contour(X, Y, Z, np.logspace(-0.5, 3, 31), cmap='gray_r')
plt.plot([c[0] for c in coords], [c[1] for c in coords])
plt.scatter(1, 1, c='red')
```

## Задача 1

Найдите градиентным спуском минимум функции $$f(x, y) = 2x^{2} - 4xy + y^{4} + 2$$

### Решение:

```python
def f(x, y):
    return 2 * x**2 - 4 * x * y + y**4 + 2


np.round(optimize.minimize(lambda x: f(*x), x0=(1, 1)).fun)
```

## Задача 2

В модели линейной регрессии три наблюдения:
- $x_{1} = 1$, $x_{2} = 3$, $x_{3} = 5$
- $y_{1} = 2.1$, $y_{2} = 2.9$, $y_{3} = 4.1$
- $\hat{y} = w_{0} + w_{1}x$

$$L = (2.1 - w_{0} - w_{1})^{2} + (2.9 - w_{0} - 3w_{1})^{2} + (4.1 - w_{0} - 5w_{1})^{2}$$

Обучите линейную регрессию при помощи градиентного спуска. Чему равно оптимальное значение $L^{2} - loss$?

```python
def f(x, y):
    return (2.1 - x - y)**2 + (2.9 - x - 3 * y)**2 + (4.1 - x - 5 * y)**2


np.round(optimize.minimize(lambda x: f(*x), x0=(1, 1)).fun)
```

## Задача 3

Найдите градиентным спуском минимум функции $$f(x, y, z) = x^{3} - 2x^{2} + y^{2} + z^{2} - 2xy + xz - yz + 3z$$

```python
def f(x, y, z):
    return x**3 - (2 * x**2) + y**2 + z**2 - (2*x*y) + (x*z) - (y*z) + (3*z)


np.round(optimize.minimize(lambda x: f(*x), x0=(0, 0, 0)).fun)
```

# Градиентный спуск с momentum

$$x^{(n+1)} = x^{(n)} - \gamma \nabla f(x^{(n)}) + \alpha(x^{(n)} - x^{(n-1)})$$

## Задача 1

Напишите следующую точку градиентного спуска с momentum для функции:

$$f(x, y, z) = 2x^{2} 4xz + 4y^{2} - 8yz + 9z^{2} + 4x + 8y - 20z$$

если текущая точка (1, 2, -5), перед ней была (0, 0, 0), $\gamma = 0.25$, $\alpha = 1$.

```python
def f(x, y, z):
    return (2 * x**2 * 4*x*z) + (4 * y**2) - (8*y*z) + (9*z**2) + 4*x + 8*y - 20*z


def grad(x, y):
    dx = 2 * x - 2 + 100 * (-4 * y * x + 4 * x ** 3)
    dy = 100 * (2 * y - 2 * x ** 2)
    return (dx, dy)
```

```python
def f(x, y, z):
    return 2*(x**2) - 4*x*z + 4*(y**2) - 8*y*z + 9*(z**2) + 4*x + 8*y - 20*z


def grad(x, y, z):
    dx = 4*x - 4*z + 4
    dy = 8*y - 8*z + 8
    dz = -4*x - 8*y + 18*z - 20
    return (dx, dy, dz)


x0 = (0, 0, 0)
gamma = 0.25
x_cur = (1, 2, -5)
alpha = 1
vals = []
coords = []
x_new = (
    x_cur[0] - gamma * grad(*x_cur)[0] + alpha * (x_cur[0] - x0[0]),
    x_cur[1] - gamma * grad(*x_cur)[1] + alpha * (x_cur[1] - x0[1]),
    x_cur[2] - gamma * grad(*x_cur)[2] + alpha * (x_cur[2] - x0[2])
)

vals.append(f(*x_cur))
coords.append(x_cur)

print(f"iter={i}; x=({x_new[0]:.2f}, {x_new[1]:.2f}, {x_new[2]:.2f});"
      f" f(x)={f(*x_new):.2f}; grad f(x)=({grad(*x_new)[0]:.2f}, {grad(*x_new)[1]:.2f}, {grad(*x_new)[2]:.2f})")
```
