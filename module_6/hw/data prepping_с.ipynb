```python
import re
import datetime
```

```python
pattern_year = re.compile('\d+\s[годалт]+')
```

```python
pattern_month = re.compile('\d+\s[мсяцв]+')
```

```python
pattern_digit = re.compile('\d+')
```

```python
test['year'] = test['ownership_period'].apply(lambda x: str(pattern_year.findall(str(x))))
test['month'] = test['ownership_period'].apply(lambda x: str(pattern_month.findall(str(x))))
test['year'] = test['year'].apply(lambda x: str(pattern_digit.findall(str(x))))
test['year'] = test['year'].apply(lambda x: x[2:-2])
test['month'] = test['month'].apply(lambda x: str(pattern_digit.findall(str(x))))
test['month'] = test['month'].apply(lambda x: x[2:-2])
```

```python
test['year'] = test['year'].apply(lambda x: 0 if x == '' else int(x))
test['month'] = test['month'].apply(lambda x: 0 if x == '' else int(x))
```

```python
test['ownership'] = test.apply(lambda x: (x['year']*12) + x['month'], axis=1)
```
