```python
import re
import datetime
```

```python
pattern = re.compile('((\d+) (года|год|лет)){0,1}( и ){0,1}((\d+) (месяц)){0,1}')
```

```python
def get_months_from_ownership_period(x):
    search_res = pattern.search(str(x))
    
    years = 0
    months = 0
    
    if search_res.group(2):
        years = int(search_res.group(2))
    if search_res.group(6):
        months = int(search_res.group(6))
    
    return years*12 + months
```

```python
test['ownership_period'] = test['ownership_period'].apply(get_months_from_ownership_period)
```
